```python
import json
import re
from wordcloud import WordCloud
import matplotlib.pyplot as plt

stop_words = ['diabetes']
taglists = []
tag_string = ''

with open('../diabetes_tweets.json', encoding="utf8") as f:
    for row in f:
        jo = json.loads(row)
        taglists.append(re.findall(r"#(\w+)", jo['text']))
      
for list in taglists:
    for tag in list:
        tag_string = tag_string + tag + ' '
            
wordcloud=WordCloud(width=1000,height=1000, background_color='white', stopwords=stop_words, min_font_size=10).generate(tag_string)

plt.figure(figsize=(10,10),facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
